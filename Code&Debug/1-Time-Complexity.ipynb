# Time Complexity

Rate of increase in time with respect to input size 

TC -> Big-oh Notation -> O( Your time taken )
1. Always calculate TC in terms of worst case
2. Avoid the constant values
3. Avoid lower bound

# Space Complexity

Memory Space -> Big Oh Notation 

Auxiliary Space + Input Space
1. Auxiliary Space -> The extra space used to solve the problem
2. Input space -> The space used to store the input
