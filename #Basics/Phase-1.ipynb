```python
# 5. Check if a given year is a leap year
# A year is a leap year if:
# It is divisible by 400
# → Leap year
# OR it is divisible by 4 AND not divisible by 100
# → Leap year
# Otherwise, it is not a leap year.

def leap_year(num):
    if not isinstance(num, int):
        'Enter a valid number'
    return 'Leap year' if((num % 400 == 0) or ((num % 4 == 0) and (num % 100 != 0))) else 'Not a Leap Year'

leap_year(2024)
```

```python
# 9. Take a character and check if it’s a vowel or consonant.

def vol_const(c):
    if not c.isalpha() and len(c) != 1:
        return 'Invalid character'
    
    c = c.lower()
    if c in ['a', 'e', 'i', 'o', 'u']:
        return 'Vowel'
    else:
        return 'Consonant'
    
vol_const('d')
```

```python
# 10. Take a character and check whether it’s uppercase, lowercase, a digit, or a special character.

def check_char_ascii(st):
    ascii_st = ord(st)
    
    if 65 <= ascii_st <= 90:
        return 'Uppercase'
    elif 97 <= ascii_st <= 122:
        return 'Lowercase'
    elif 48 <= sacii_st <= 57:
        return 'Digit'
    else:
        return 'Special Character'
    
check_char_ascii('s')
```
