# Problem 1: Two Sum

```python
# Brute Force Approach
# Time Complexity : O(n^2)
# Space Complexity : O(n)

def twosum_bf(arr, target):
    n = len(arr)
    for i in range(n):
        for j in range(i+1, n):
            if arr[i] + arr[j] == target:
                return [i, j]

arr = [3, 1, 4, 6, 5]
target = 9
twosum_bf(arr, target)
```

```python
# Better Approach - HashMap
# Core Idea
# â€œInstead of checking every pair, we can reduce repeated work by remembering what weâ€™ve already seen.â€

# Time Complexity : O(n^2)
# Space Complexity : O(n)

def twosum_ba(arr, target):
    seen = {}   #value -> index
    
    for i, x in enumerate(arr):
        need = target - x
        if need in seen:
            return [seen[need], i]
        seen[x] = i
        
        print(i,x,need,seen)
        
arr = [3, 1, 4, 6, 5]
target = 9
twosum_ba(arr, target)
```

Final Interview-Quality Answer (Memorize this)

â€œThe brute-force solution checks all pairs in O(nÂ²).
To optimize, I use a hash map.
While traversing the array, for each number I calculate its complement with respect to the target.
If that complement already exists in the map, I return the stored index and the current index.
Otherwise, I store the current number with its index.
This gives O(n) time complexity with O(n) extra space.â€

# Problem 2: Best Time to Buy and Sell Stock


```python
# Brute Force Approach - Evaluate all buy-sell pairs
# Time Complexity : O(n^2)
# Space Complexity : O(1)

def buy_sell_bf(prices):
    n = len(prices)
    max_profit = 0
    
    for i in range(n):  #choose buy day
        for j in range(i+1, n): #choose sell day after buy day
            profit = prices[j] - prices[i]
            if profit > max_profit:
                max_profit = profit
                
    return max_profit
    

prices = [7,1,5,3,6,4]
buy_sell_bf(prices)
```

```python
# Better Approach
# Time Complexity: O(n)
# Space Complexity : O(1)

def buy_sell_ba(prices):
    min_price = float('inf')
    max_profit = 0
    
    for i in prices:
        if i < min_price:
            min_price = i
        else:
            profit = i - min_price
            if profit > max_profit:
                max_profit = profit
    return max_profit

prices = [7,1,5,3,6,4]
buy_sell_bf(prices)
```

```python
# OPtimal Approach

def buy_sell_oa(prices):
    min_price = float('inf')
    max_profit = 0
    for i in prices:
        min_price = min(min_price, i)
        max_profit = max(max_profit, i - min_price)
        
    return max_profit

prices = [7,1,5,3,6,4]
buy_sell_oa(prices)
```

# Problem = Max Subarray

```python
def max_subarray_bf(nums):
    n = len(nums)
    maxsum = float('-inf')
    for i in range(n):
        currentsum = 0
        for j in range(i+1, n):
            currentsum = currentsum + nums[j]
            maxsum = max(maxsum, currentsum)
            
    return maxsum
    
    
nums = [-2,1,-3,4,-1,2,1,-5,4]
max_subarray_bf(nums)
```

```python
def max_subarray_oa(nums):
    currentsum = nums[0]
    maxsum = nums[0]
    
    for i in range(1, len(nums)):
        currentsum = max(nums[i], currentsum + nums[i])
        maxsum = max(maxsum, currentsum)
        
    return maxsum

nums = [-2,1,-3,4,-1,2,1,-5,4]
max_subarray_oa(nums)
```

# Problem Statement (Easy â†’ Mid)
Two Sum (Classic Array Problem)
Given an array of integers nums and an integer target, return the indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution.

You may not use the same element twice.

```python
# Brute Force Approach
# Time Complexity : O(n^2)
# Space Complexity : O(n)

def two_sum_bf(arr, target):
    n = len(arr)
    for i in range(n):
        for j in range(i+1,n):
            if (arr[i] + arr[j] == target):
                return [i, j]

arr = [3, 1, 4, 6, 5]
target = 9
two_sum_bf(arr, target)
```

```python
# Better Approach with sortinf and two pointers
# Time Complexity : O(nlogn) due to sorting
# Space Complexity : O(n) > if we store index mapping

def two_sum_ba(arr, target):
    index_arr = list(enumerate(arr))
    index_arr.sort(key=lambda x: x[1])
    left = 0
    right = len(arr)-1
    
    while left < right:
        curr_sum = index_arr[left][1] + index_arr[right][1]
        
        if curr_sum == target:
            return [index_arr[left][0], index_arr[right][0]]
        elif curr_sum < target:
            left += 1
        else:
            right -= 1
            
    return []

arr = [3, 1, 4, 6, 5]
target = 9
two_sum_ba(arr, target)
```

```python
# Optimal Approach
# Time complexity = O(n) because we scan once.
# Space complexity = O(n) for the hash map.

def two_sum_oa(arr, target):
    seen = {}
    for i, j in enumerate(arr):
        complement = target - j
        if complement in seen:
            return [seen[complement], i]
        seen[j] = i
    return []

arr = [3, 1, 4, 6, 5]
target = 9
two_sum_oa(arr, target)
```

# Problem 2: "Best Time to Buy and Sell Stock"

ðŸ“œ Statement:
Tujhe ek array diya hai prices, jisme prices[i] ka matlab hai ki i-th din ka stock ka price kya hai.
Tujhe sirf ek baar buy aur ek baar sell karna allowed hai (buy pehle aur sell baad me honi chahiye).
Maximum profit kitna ho sakta hai ye batana hai.
Agar profit nahi ban sakta to 0 return karo.

Input: prices = [7, 1, 5, 3, 6, 4]
<br>Output: 5  (Buy at 1, sell at 6)


```python
# Brute Force Approach
# Time Complexity: O(nÂ²)

def buy_sell_bf(arr):
    n = len(arr)
    max_profit = 0
    
    for i in range(n):  #Buy day
        for j in range(i+1, n):  #Sell day
            profit = arr[j] - arr[i]
            if profit > max_profit:
                max_profit = profit
            
    return max_profit

arr = [7, 1, 5, 3, 6, 4]
buy_sell_bf(arr)
```

```python
# Better Approach
# Time Complexity: O(nÂ²)

def buy_sell_ba(arr):
    minPrice = float('inf')
    maxProfit = 0
    n = len(arr)
    for i in range(n):
        if arr[i] < minPrice:
            minPrice = arr[i]
            
        profit = arr[i] - minPrice
        if profit > maxProfit:
            maxProfit = profit
            
    return maxProfit

arr = [7, 1, 5, 3, 6, 4]
buy_sell_ba(arr)
```

```python
# Optimal Approach
# O(n) time & O(1) space.

# "Minimum pakad, Maximum soch"
# Minimum = cheapest buy
# Maximum = best sell profit

def buy_sell_oa(arr):
    min_Price = float('inf')
    max_Profit = 0
    
    for price in arr:
        min_Price = min(price, min_Price)
        max_Profit = max(max_Profit, price - min_Price)
        
    return max_Profit


arr = [8, 4, 2, 10, 6]
buy_sell_oa(arr)
```

# Problem 1: Running Sum of 1D Array.

```python
# Brute Force 
# Complexity:
# Time: O(nÂ²) (because of the nested loop).
# Space: O(n) (for res).


def runningSum(nums):
    res = []
    for i in range(len(nums)):
        total = 0
        for j in range(i+1):
            total += nums[j]
            
        res.append(total)
    return res

print(runningSum([1,2,3,4]))
```

```python
def runningSum2(nums):
    res = [0] * len(nums)
    res[0] = nums[0]
    for i in range(1, len(nums)):
        res[i] = res[i-1] + nums[i]
        
    return res

print(runningSum2([1,2,3,4]))
```

# Problem 2: Maximum Subarray

```python
# Brute Force 
# Time: O(nÂ³) (3 loops)
# Space: O(1)

def maxSubArray(nums):
    max_sum = float('-inf')
    for i in range(len(nums)):
        for j in range(i, len(nums)):
            current_sum = 0
            for k in range(i, j+1):
                current_sum += nums[k]
            max_sum = max(max_sum, current_sum)
    return max_sum

maxSubArray([1,-2,2,4,-3,1,2])
```

```python
def maxSubarray_bf(nums):
    n = len(nums)
    max_sum = float('-inf')
    for i in range(n):
        for j in range(i, n):
            curr_sum = sum(nums[i:j+1])
            max_sum = max(max_sum, curr_sum)
            
    return max_sum

maxSubarray_bf([1,-2,2,4,-3,1,2])
```

```python
# Better Approach (O(nÂ²)) >> (Prefix Sum / Running Sum)
# Time: O(nÂ²) (2 loops)
# Space: O(1)
    
def maxSubArray_ba(nums):
    max_sum = float('-inf')
    for i in range(len(nums)):
        current_sum = 0
        for j in range(i, len(nums)):
            current_sum += nums[j]
            max_sum = max(max_sum, current_sum)
    return max_sum

maxSubArray_ba([1,-2,2,4,-3,1,2])
```

```python
# Optimal Approach â€“ Kadaneâ€™s Algorithm (O(n))

def maxSubArray_oa(nums):
    max_ending_here = nums[0]
    max_so_far = nums[0]
    
    for i in range(1, len(nums)):
        max_ending_here = max(nums[i], nums[i] + max_ending_here)
        max_so_far = max(max_so_far, max_ending_here)
        
    return max_so_far

maxSubArray_oa([1,-2,2,4,-3,1,2])
```

```python
def maxSubArray_oa2(nums):
    curr_sum = max_sum = nums[0]

    for i in nums[1:]:
        curr_sum = max(i, curr_sum+i)
        max_sum = max(curr_sum, max_sum)

    return max_sum

maxSubArray_oa2([1,-2,2,4,-3,1,2])
```

# Problem: Subarray Sum Equals K

```python
# Brute Force (O(nÂ³))
# Time: O(nÂ³)
# Space: O(1)

def subarraysum_bf(nums, k):
    count = 0
    n = len(nums)
    for i in range(n):
        for j in range(i, n):
            total = 0
            for t in range(i, j+1):
                total += nums[t]
            if total == k:
                count += 1
                
    return count

subarraysum_bf([1,-2,2,4,-3,1,2], 4)
```

```python
# Better Approach (O(nÂ²))
# Time: O(n2)
# Space: O(1)

def subsumarray_ba(nums, k):
    count = 0
    n = len(nums)
    for i in range(n):
        current_sum = 0
        for j in range(i, n):
            current_sum += nums[j]
            if current_sum == k:
                count += 1
    return count

subsumarray_ba([1,-2,2,4,-3,1,2], 6)
```

```python
# Optimal Approach (O(n)) â€“ HashMap
# Time: O(n)
# Space: O(n) (hashmap)

def subsumarray_oa(nums, k):
    count = 0
    n = len(nums)
    sum_freq = {0: 1}  # prefix sum 0 appears once
    current_sum = 0
    
    for i in nums:
        current_sum += i
        if current_sum - k in sum_freq:
            count += sum_freq[current_sum - k]
        sum_freq[current_sum] = sum_freq.get(current_sum, 0)  + 1
        
    return count

subsumarray_oa([1,-2,2,4,-3,1,2], 6)
```

# Subarray Sums Divisible by K

```python
# Brute Force 
# O(nÂ²).

def subarrayDivbyk(nums, k):
    count = 0
    n = len(nums)
    for i in range(n):
        current_sum = 0
        for j in range(i, n):
            current_sum += nums[j]
            if current_sum % k == 0:
                count += 1
                
    return count
    
    
subarrayDivbyk([4,5,0,-2,-3,1], 5)
```

```python
# Optimal Approach (O(n)) â€“ Prefix Sum + HashMap
# Analytical insight:

# A subarray (i..j) is divisible by k if:
# (prefix[j] - prefix[i]) % k == 0

# Equivalently:
# prefix[j] % k == prefix[i] % k.


def subarrayDivbyk_oa(nums, k):
    count = 0
    prefix_sum = 0
    remainder_freq = {0: 1}
    
    for i in nums:
        prefix_sum += i
        remainder = prefix_sum % k
        
        if remainder < 0:
            remainder += k
            
        if remainder in remainder_freq:
            count += remainder_freq[remainder]
            
            
```

# Problem 3 â€” Find Pivot Index

Given an array of integers nums, return the pivot index of this array.

The pivot index is the index where the sum of all numbers strictly to the left is equal to the sum of all numbers strictly to the right.
If no such index exists, return -1. If there are multiple, return the left-most pivot.

```python
# Brute Force Approach
# Complexity:
# Time: O(nÂ²) (nested summations)
# Space: O(1)

def pivotIndex_bf(arr):
    n = len(arr)
    
    for i in range(n):
        left_sum = sum(arr[:i])
        right_sum = sum(arr[i+1:])
        if left_sum == right_sum:
            return i
    return -1
    
print(pivotIndex_bf([1,7,3,6,5,6]))
print(pivotIndex_bf([1,2,3]))
```

```python
# Better Approach
# Complexity:
# Time: O(n)
# Space: O(n)

def pivotIndex_ba(arr):
    n = len(arr)
    
    prefix = [0]*n
    prefix[0] = arr[0]
    
    for i in range(1, n):
        prefix[i] = prefix[i-1] + arr[i]
        
    total_sum = prefix[-1]
    
    for i in range(n):
        left_sum = prefix[i-1] if i > 0 else 0
        right_sum = total_sum - prefix[i]
        if left_sum == right_sum:
            return i
        
    return -1
    
print(pivotIndex_bf([1,7,3,6,5,6]))
print(pivotIndex_bf([1,2,3]))
```

```python
# Optimal Appraoch - Optimized Prefix Sum (Single Pass)

def pivotIndex_oa(arr):
    total_sum = sum(arr)
    left_sum = 0
    
    for i, num in enumerate(arr):
        if left_sum == (total_sum - left_sum - num):
            return i
        left_sum += num
    return -1
        
print(pivotIndex_bf([1,7,3,6,5,6]))
print(pivotIndex_bf([1,2,3]))
```

# Problem 4 â€” Product of Array Except Self

```python
# Brute Force
# Complexity:
# Time = O(nÂ²)
# Space = O(1) extra

def product_except_self_bruteforce(nums):
    n = len(nums)
    result = []
    
    for i in range(n):
        prod = 1
        for j in range(n):
            if i != j:
                prod *= nums[j]
        result.append(prod)
        
    return result 

print(product_except_self_bruteforce([1,2,3,4]))
print(product_except_self_bruteforce([-1,1,0,-3,3]))
```

```python
# Better Approach >> Prefix & Suffix Arrays
# Time = O(n)
# Space = O(n)

def product_except_self_prefix_suffix(nums):
    n = len(nums)
    prefix = [1] * n
    suffix = [1] * n
    result = [1] * n
    
    for i in range(1,n):
        prefix[i] = prefix[i-1] * nums[i-1]
        
    for i in range(n-2, -1, -1):
        suffix[i] = suffix[i+1] * nums[i+1]
        
    for i in range(n):
        result[i] = prefix[i] * suffix[i]
        
    return result


print(product_except_self_prefix_suffix([1,2,3,4]))
print(product_except_self_prefix_suffix([-1,1,0,-3,3]))
```

```python
# Optimal Approach
# Time = O(n)
# Space = O(1) extra (ignoring output array)

def product_except_self(nums):
    n = len(nums)
    result = [1]*n
    
    prefix = 1
    for i in range(n):
        result[i] = prefix
        prefix *= nums[i]
        
    suffix = 1
    for i in range(n-1, -1, -1):
        result[i] *= suffix
        suffix *= nums[i]
        
    return result

print(product_except_self([1,2,3,4]))
print(product_except_self([-1,1,0,-3,3]))
```
