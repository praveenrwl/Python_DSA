# Problem Statement:
"Two Sum"
Tujhe ek array aur ek target diya gaya hai. Tera kaam hai do alag indexes dhoondhna jinka sum exactly target ke equal ho.
Return karo un dono indexes ka list.

```python
# Brute Force Approach
# Time complexity: O(nÂ²).

def two_sum(arr, target):
    n = len(arr)
    for i in range(n):
        for j in range(i+1, n):
            if arr[i] + arr[j] == target:
                return arr[i], arr[j]
            
nums = [2, 5, 7, 9, 10, 11, 15]
target = 21
two_sum(nums, target)
```

```python
# Better Approach >> Hashmap
# Time complexity: O(n)

def two_sum_ba(arr, target):
    temp = {}
    n = len(arr)
    for i in range(n):
        complement = target - arr[i]
        if complement in temp:
            return [temp[complement], i]
        temp[arr[i]] = i
    return []
        
    
nums = [2, 5, 7, 9, 10, 11, 15]
target = 21
two_sum_ba(nums, target)
```

```python
# Optimal Approach >> Hashmap
# Time complexity: O(n)


def two_sum_oa(arr, target):
    arr = [(num, i) for i, num in enumerate(nums)]  # Make pairs
    arr.sort()  #sort by num
    left, right = 0, len(arr)-1
    
    while left < right:
        curr_sum = arr[left][0] + arr[right][0]
        if curr_sum == target:
            return [arr[left][1], arr[right][1]]
        elif curr_sum < target:
            left += 1
        else:
            right -= 1
    return []

nums = [2, 5, 7, 9, 10, 11, 15]
target = 21
two_sum_ba(nums, target)
    
```

# Problem 1: 3Sum

| Metric | Complexity | Explanation               |
| ------ | ---------- | ------------------------- |
| Time   | **O(nÂ³)**  | 3 nested loops            |
| Space  | **O(m)**   | m = no. of valid triplets |



```python
def threeSum_bruteforce(nums):
    n = len(nums)
    result = []
    
    for i in range(n):
        for j in range(i+1, n):
            for k in range(j+1, n):
                if nums[i] + nums[j] + nums[k] == 0:
                    triplet = sorted([nums[i], nums[j], nums[k]])
                    if triplet not in result:
                        result.append(triplet)
                    
    return result
                
arr = [-1, 0, 1, 2, -1, -4]
threeSum_bruteforce(arr)
                
```

```python
# Better Approach: Two Pointer Approach
# | Metric | Value     | Explanation                     |
# | ------ | --------- | ------------------------------- |
# | Time   | **O(nÂ²)** | Outer loop * (two-pointer O(n)) |
# | Space  | **O(1)**  | Ignoring output list            |


def threesum_ba(nums):
    nums.sort()
    res = []
    
    for i in range(len(nums)):
        if i > 0 and nums[i] == nums[i-1]:
            continue
            
        left, right = i+1, len(nums)-1

        while left < right:
            total = nums[i] + nums[left] + nums[right]
            
            if total < 0:
                left += 1
            elif total > 0:
                right -= 1
            else:
                res.append([nums[i], nums[left], nums[right]])
                
                left += 1
                right -= 1
                
                while left < right and nums[left] == nums[left-1]:
                    left += 1
                    
                while left < right and nums[right] == nums[right + 1]:
                    right -= 1
    return res
            
nums = [-1, 0, 1, 2, -1, -4]
threesum_ba(nums)
```

```python
# ðŸ’­ 1ï¸âƒ£ Hash-based Approach (HashSet + HashMap idea)

def threesum_oa(nums):
    result = set()
    nums.sort()
    
    for i in range(len(nums)):
        target = -nums[i]
        seen = set()
        for j in range(i+1, len(nums)):
            complement = target - nums[j]
            if complement in seen:
                result.add((nums[i], nums[j], complement))
            else:
                seen.add(nums[j])
                
    return result

threesum_oa(nums)
```

# Container With Most Water

```python
# Brute Force 
# Time: O(nÂ²)
# Space: O(1)

def maxArea_bruteforce(height):
    max_water = 0
    n = len(height)
    for i in range(n):
        for j in range(i+1, n):
            h = min(height[i], height[j])
            w = j - i 
            max_water = max(max_water, h*w)
    
    return max_water

maxArea_bruteforce([1,8,6,2,5,4,8,3,7])
```

```python
# Better Approach
def maxArea_better(height):
    max_water = 0
    n = len(height)
    
    for i in range(n):
        for j in range(n-1, i, -1):
            # max possible area from this pair
            possible = min(height[i], height[j]) * (j-i)
            
            if possible < max_water:
                break
                
            max_water = max(max_water, possible)
            
    return max_water

maxArea_better([1,8,6,2,5,4,8,3,7])
```
